```python
def validation_check_non_unique(dataFrame, col, sId,status_df, filename = 'Unknow_file'):
    '''
    check for empty value in the dataframe with column
    '''
    dataFrame = dataFrame[dataFrame.duplicated(subset=[col])]
    row_status = {}
    print(dataFrame)
    if not dataFrame.empty:
        for index, row in dataFrame.iterrows():
                desc = ' FileName_ '+filename + ' RowNumber_' + str(index)
                row_status = {'id': row[sId], 'action': 'Rejected_Non_Unique','description' : desc} 
                status_df = status_df.append(row_status,ignore_index=True)
        return status_df
    return
        
    
```

```python
def normalize_validation_date_check(dataFrame, col, sId,status_df, filename = 'Unknow_file'):
    '''
    check for valid dates in the dataframe on date
    '''
    df = dataFrame.copy(deep=True)
    dataFrame[col] = pd.to_datetime(dataFrame[col].dt.date,errors='coerce')
    row_status = {}
    df_rejected = df[dataFrame[col].isnull()]
    dataFrame = dataFrame[np.invert(dataFrame[col].isnull())]
    if not df_rejected.empty:
        for index, row in df_rejected.iterrows():
                desc = 'FileName_ '+filename + ' RowNumber_' + str(index)
                row_status = {'id': row[sId], 'action': 'Rejected - INVALID DATE','description' : desc} 
                status_df = status_df.append(row_status,ignore_index=True)
        return status_df, dataFrame
    return status_df, dataFrame
```
